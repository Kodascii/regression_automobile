```python
df = pd.read_csv("train.csv")

train = clean_df(df, 0)
train_30 = clean_df(df, 30)
train_50 = clean_df(df, 50)

X = train.drop('Price', axis=1)
y = train['Price']

X_train, X_test, y_train, y_test = train_test_split(X,  y, random_state=42, test_size=0.2)
print(X_train.shape)
print(y_train.shape)
```

```python
param_grid = {"gradientboostingregressor__n_estimators" : [10, 100]}
 #             "gradientboostingresgressor_epsilon" : [5,10,20]}

# param_grid = {"n_estimators" : [10, 100]}

kf = KFold(n_splits=5, shuffle=True, random_state=0)


svr_cv = GridSearchCV(model, param_grid = param_grid, cv=5, scoring="neg_mean_squared_error")
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score, mean_squared_error

# Configuration de KFold
kf = KFold(n_splits=5, shuffle=True, random_state=0)

# Initialiser des listes pour enregistrer les scores pour chaque pli
mse_scores = []
rmse_scores = []
r2_scores = []



# Boucler sur chaque pli défini par KFold
for train_index, test_index in kf.split(X):

    # On redéfini les données d'apprentissage et de test pour chaque pli
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    

    svr_cv.fit(X_train, y_train)
    y_pred = svr_cv.predict(X_test)

    # Calcul du MSE
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)
    
    # Calcul du RMSE
    rmse = np.sqrt(mse)
    rmse_scores.append(rmse)
    
    # Calcul du R2
    r2 = r2_score(y_test, y_pred)
    r2_scores.append(r2)

print(f"MSE : {mse_scores}")
print(f"RMSE : {rmse_scores}")
print(f"R² : {r2_scores}")

print("Mean MSE:", np.mean(mse_scores))
print("Mean RMSE:", np.mean(rmse_scores))
print("Mean R²:", np.mean(r2_scores))
```
