```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from csv_cleaner import clean_df
from pipeline import pipeline_create

def load_and_clean_data(file_path, cleaning_threshold):
    """Étape 1 : Charger les données et les nettoyer."""
    df = pd.read_csv(file_path)
    df_cleaned = clean_df(df, cleaning_threshold)
    return df_cleaned

def calculate_correlation_matrix(df):
    """Étape 2 : Calculer la matrice de corrélation."""
    # Sélectionner uniquement les colonnes numériques pour la matrice de corrélation
    df_numeric = df.select_dtypes(include=[np.number])
    return df_numeric.corr()


def visualize_correlation_matrix(corr_matrix):
    """Étape 3 : Visualiser la matrice de corrélation."""
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
    plt.title('Matrice de corrélation')
    plt.show()

def data_analysis_pipeline(file_path, cleaning_threshold):
    """Pipeline pour l'analyse de données."""
    # Étape 1 : Charger les données et les nettoyer
    df_cleaned = load_and_clean_data(file_path, cleaning_threshold)
    
    # Étape 2 : Calculer la matrice de corrélation
    corr_matrix = calculate_correlation_matrix(df_cleaned)
    
    # Étape 3 : Visualiser la matrice de corrélation
    visualize_correlation_matrix(corr_matrix)

# Utiliser la pipeline
data_analysis_pipeline('../res/train.csv', 10)
```
